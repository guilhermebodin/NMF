## 0_Imports e configs

```python
import pandas as pd
```

## 1_Lendo dados dos estabelecimentos

```python
df_business = pd.read_json('yelp_dataset/yelp_academic_dataset_business.json',lines=True,)
df_business = df_business[['business_id','name','city','state','stars','review_count','categories']]
```

##### Contando quantidade de estabelecimentos por estados 

```python
df_business.shape
```

```python
business_per_state = df_business.groupby(by='state').count()['business_id'].sort_values(ascending=False)
state_list = business_per_state.index[0:10]
business_per_state
```

##### Criando dicion√°rios para apendar as partes dos datasets dos estados

```python
state_dict_parts = {}
state_business_list = {}
for state in state_list:
    state_dict_parts[state] = []
    state_business_list[state] = df_business.loc[df_business['state']== state,'business_id'].tolist()
```

## 2_Lendo dados das reviwes

```python
df_reader = pd.read_json('yelp_dataset/yelp_academic_dataset_review.json',lines=True,chunksize=100000)
```

```python
for state in state_list:
    print(f'{state}  ',end=', ')
print()
for part in df_reader:
    for state in state_list:
        part_x = part[part['business_id'].isin(state_business_list[state])]
        part_x = part_x[['business_id','stars','useful','funny','cool','text']]
        print(part_x.shape[0],end=', ')
        state_dict_parts[state].append(part_x)
    print()
```

```python
for state in state_list:
    df_reviews_x = pd.concat(state_dict_parts[state],axis=0)
    df_reviews_x.to_csv(f'yelp_dataset/reviews/{state}_reviews.csv')
    print(f'{state}: {df_reviews_x.shape[0]}')
    
```
